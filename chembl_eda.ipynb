```python
import polars as pl

uniprot = (
    pl.read_csv("datasets/uniprot_chembl.tsv", separator="\t")
      .select(["Entry", "Sequence", "ChEMBL"])
      .with_columns(
          # (A) strip a *single* trailing â€œ;â€ then split the remaining IDs
          #     â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
          # -- for â‰¥0.19.3 use:
          # pl.col("ChEMBL").str.strip_chars_end(";")
          #
          # -- for 0.18â€“0.19 use:
          # pl.col("ChEMBL").str.rstrip(";")
          #
          # -- for <0.18 use:
          # pl.col("ChEMBL").str.strip_matches(";")
          pl.col("ChEMBL").str.strip_chars_end(";")  # â† pick the right one here
            .str.split(";")
            .alias("chembl_list")
      )
      .explode("chembl_list")        # one row per individual ID
      .drop(["ChEMBL"])
)
uniprot
```
